# Box Plots
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')
```

```python
df = pd.read_csv('players_21.csv')
df.head()
```

```python
df.shape
```

```python
fig = plt.figure(figsize=(5,5,))
ax = fig.add_axes([0,0,1,1])

ax.set_title('Box Plot')

ax.boxplot(data=df, x='overall', labels=['overall score'])

plt.show()
```

### Removing Outliers with the `showfliers=`

```python
fig = plt.figure(figsize=(5,5,))
ax = fig.add_axes([0,0,1,1])

ax.set_title('Box Plot')

ax.boxplot(data=df, x='overall', labels=['overall score'], 
          showfliers=False)

plt.show()
```

### Setting `vert=False`

```python
fig = plt.figure(figsize=(5,5,))
ax = fig.add_axes([0,0,1,1])

ax.set_title('Box Plot')

ax.boxplot(data=df, x='overall', labels=['overall score'], 
          showfliers=False, vert=False)

plt.show()
```

## Data Preparation to create 2 DataFrames for 2 seperate teams

```python
inter = df[df['club_name'] == 'Inter']['overall']
```

```python
baca = df[df['club_name'] == 'FC Barcelona']['overall']
```

```python
data = [inter, baca]
```

```python
type(data)
```

## Comparing two different teams with box plots

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0,0,1,1])

ax.boxplot(data)

plt.show()
```

```python
fig = plt.figure(figsize=(5,5))
ax = fig.add_axes([0,0,1,1])
ax.set_title('Box Plot')

ax.boxplot(data, labels=['Inter Milan', 'FC Barcelona'])

plt.show()
```

# Violin Plots
https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.violinplot.html

```python
fig = plt.figure(figsize=(5,5))

ax = fig.add_axes([0,0,1,1])
ax.set_title('Violin Plots')
ax.violinplot(data)

plt.show()
```

## Adding `xticks`

```python
fig = plt.figure(figsize=(5,5))

ax = fig.add_axes([0,0,1,1])
ax.set_title('Violin Plots')
ax.violinplot(data)

plt.xticks([1,2],['Inter Milan', 'FC Barcelona'], rotation=45)

plt.show()
```

```python
fig = plt.figure(figsize=(5,5))

ax = fig.add_axes([0,0,1,1])
ax.set_title('Violin Plots')
ax.set_xlabel('Team')
ax.set_ylabel('Overall')
ax.violinplot(data)
ax.boxplot(data)

plt.xticks([1,2],['Inter Milan', 'FC Barcelona'])

plt.show()
```
