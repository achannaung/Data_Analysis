# Scatter Plot

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')  # Optional: for ggplot-like style
```

```python
df = pd.read_csv('players_21.csv')
df.head()
```

```python
df.shape
```

```python
players_100 = df.head(100)
```

```python
df.columns.tolist()
```

```python
players_100[['overall', 'wage_eur']].head()
```

# Scatter Plot
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html

```python
fig = plt.figure(figsize=(10,10))

ax1  = fig.add_axes([0,0,1,1,])
ax1.set_title('Chart')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

ax1.scatter(data=players_100, x ='overall', y='wage_eur')
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))

ax1  = fig.add_axes([0,0,1,1,])
ax1.set_title('Overall ability Vs Wages')
ax1.set_xlabel('Overall ability')
ax1.set_ylabel('Wages')

ax1.scatter(data=players_100, x ='overall', y='wage_eur', s=30)
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))

ax1  = fig.add_axes([0,0,1,1,])
ax1.set_title('Overall ability Vs Wages')
ax1.set_xlabel('Overall ability')
ax1.set_ylabel('Wages')

ax1.scatter(data=players_100, x ='overall', y='wage_eur', s=players_100['value_eur']/150000, color='crimson')
plt.show()
```
