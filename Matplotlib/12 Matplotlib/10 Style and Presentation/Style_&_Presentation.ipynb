```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.style.use('ggplot')
```

```python
x = np.arange(0,10,0.1)
x
```

```python
x.shape
```

```python
y1 = x
y2 = x**2
y3 = x**3
```

```python
(y1.shape, y2.shape, y3.shape)
```

## Labels & Font size

#### Links and resources:
* https://matplotlib.org/stable/api/_as_gen/matplotlib.axes.Axes.set_title.html
* https://matplotlib.org/stable/gallery/pyplots/fig_axes_labels_simple.html
* https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xticks.html
* https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.yticks.html

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0,0,1,1])

ax.plot(x, y1)
ax.plot(x, y2)
ax.plot(x, y3)

ax.set_xlim(0,10)
ax.set_ylim(0,100)

plt.show()
```

## Add Title & Label

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0,0,1,1])

ax.plot(x, y1)
ax.plot(x, y2)
ax.plot(x, y3)

ax.set_xlim(0,10)
ax.set_ylim(0,100)

ax.set_title('Chart Title Left', loc='left')
ax.set_title('Chart Title Center', loc='center')
ax.set_title('Chart Title Right', loc='right')

ax.set_xlabel("X Label")
ax.set_ylabel("Y Label")
plt.show()
```

## Change Font

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0,0,1,1])

ax.plot(x, y1)
ax.plot(x, y2)
ax.plot(x, y3)

ax.set_title('Chart Title', 
            fontsize=40, color='red',fontname='Impact')

ax.set_xlabel("X Label", 
             fontname='times', rotation='vertical')
ax.set_ylabel("Y Label",
             fontname='Tahoma')
plt.show()
```

## Change `xticks`

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0,0,1,1])

ax.plot(x, y1)
ax.plot(x, y2)
ax.plot(x, y3)

ax.set_title('Chart Title', fontsize=40, color='red',fontname='Impact')

ax.set_xlabel("X Label",fontname='times', rotation='vertical')
ax.set_ylabel("Y Label",fontname='Tahoma')

plt.xticks([0,2,4,6,8,10],['zero', 'two', 'four', 'six', 'eight', 'ten'])

plt.show()
```

### `xicks` rotation

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0,0,1,1])

ax.plot(x, y1)
ax.plot(x, y2)
ax.plot(x, y3)

ax.set_title('Chart Title', fontsize=40, color='red',fontname='Impact')

ax.set_xlabel("X Label",fontname='times', rotation='vertical')
ax.set_ylabel("Y Label",fontname='Tahoma')

plt.xticks([0,2,4,6,8,10],['zero', 'two', 'four', 'six', 'eight', 'ten'],
          rotation='vertical')

plt.show()
```

## Legend

#### Links and resources:
* https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.legend.html

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0,0,1,1])

ax.plot(x, y1, label='y=x')
ax.plot(x, y2, label='y=x**2')
ax.plot(x, y3, label='y=x**3')

plt.legend()

plt.show()
```

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0,0,1,1])

ax.plot(x, y1, label='y=x')
ax.plot(x, y2, label='y=x**2')
ax.plot(x, y3, label='y=x**3')

plt.legend(loc='center')

plt.show()
```

## Legend with categorical plots

```python
url ='https://raw.githubusercontent.com/achannaung/Data_Visualization_with_Python/main/Udemy/top_10_countries.csv?token=GHSAT0AAAAAAB5R2HGUFAR3ZNZR3RVBUHJGY6HWW6A'
df = pd.read_csv(url)
df.head(3)
```

```python
df.rename(columns={'Country / Dependency':'Country'}, inplace=True)
```

```python
df.head(2)
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_axes([0,0,1,1])

for country in df['Country']:
    ax.bar('Country', 'Population', data= df[df['Country']==country], label=country)
    
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

## Annotating

#### Links and resources:
* https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.annotate.html
* https://matplotlib.org/stable/tutorials/text/annotations.html

```python
fig = plt.figure(figsize = (5,5))
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,y1)
ax1.plot(x,y2)
ax1.plot(x,y3)

plt.annotate('Blue line',(2,200))
plt.arrow(3,190, dx=2,dy=-60, color='black')

plt.show()
```

```python
fig = plt.figure(figsize = (5,5))
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,y1)
ax1.plot(x,y2)
ax1.plot(x,y3)

plt.annotate('Blue line',
            xy=(7,350), xycoords='data',
            xytext=(0.5,0.5), textcoords='axes fraction',
            arrowprops=dict(facecolor='red', edgecolor='black'),
            horizontalalignment='left', verticalalignment='top')



plt.show()
```

## Grids

#### Links and resources:
* https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.grid.html
* https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html

```python
fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0,0,1,1])

ax.plot(x, y1, color= 'crimson')
ax.plot(x, y2, color='lime')
ax.plot(x, y3, color='purple')

plt.grid(True, color='gray', linestyle='dotted', linewidth=2)


plt.show()
```
