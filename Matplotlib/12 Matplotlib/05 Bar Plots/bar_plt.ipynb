```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('top_10_countries.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.rename(columns={'Country / Dependency':'Country'}, inplace=True)
```

```python
df.head(2)
```

### Create bar_plot population by region

```python
rp = df[['Region', 'Population']]
rp.head(2)
```

### Group By

```python
rp = rp.groupby(by='Region').sum()
```

```python
rp.reset_index(inplace=True)
```

```python
rp
```

```python
fig = plt.figure(figsize=(5,5))

axes = fig.add_axes([0,0,1,1])

axes.set_xlabel('Region')
axes.set_ylabel('Total Population')
axes.set_title('Top 10 Countries Population by Region')
```

### Adding Bar Plot

```python
fig = plt.figure(figsize=(5,5))

axes = fig.add_axes([0,0,1,1])

axes.set_xlabel('Region')
axes.set_ylabel('Total Population')
axes.set_title('Top 10 Countries Population by Region')


plt.bar(x= rp['Region'], height= rp['Population'])

plt.show()
```

## Adding Color

```python
fig = plt.figure(figsize=(5,5))

axes = fig.add_axes([0,0,1,1])

axes.set_xlabel('Region')
axes.set_ylabel('Total Population')
axes.set_title('Top 10 Countries Population by Region')


plt.bar(x= rp['Region'], height= rp['Population'], color=['y','g', 'r','b'], edgecolor='black')

plt.show()
```

```python
rp.sort_values('Population', ascending=False, inplace=True)
```
