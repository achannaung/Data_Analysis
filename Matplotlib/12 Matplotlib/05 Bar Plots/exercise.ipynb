import matplotlib.pyplot as plt

df = pd.read_csv('top_10_countries.csv')
df.head()

```python
df.shape
```

```python
df.columns
```

```python
df.rename(columns={'Country / Dependency':'Country'}, inplace=True)
```

```python
df.head(2)
```

```python
crp = df[['Region', 'Population']]
crp.head(2)
```

```python
crp = crp.groupby(by='Region').sum()
crp
```

```python
crp.reset_index(inplace=True)
```

```python
fig = plt.figure(figsize=(5,5))

axes  = fig.add_axes([0,0,1,1])
axes.set_xlabel('Region')
axes.set_ylabel('Population')
axes.set_title('Top 10 Countries Population by Region')

plt.bar(data=crp, x='Region', height='Population', color=['lime','r','purple','deeppink'])
plt.show()
```

```python
crp.sort_values(by='Population', ascending=False, inplace=True)
```

```python
crp
```

```python
fig = plt.figure(figsize=(5,5))

axes  = fig.add_axes([0,0,1,1])
axes.set_xlabel('Region')
axes.set_ylabel('Population')
axes.set_title('Top 10 Countries Population by Region')

plt.bar(data=crp, x='Region', height='Population', color=['r','purple','deeppink', 'navy'])
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,8))

p1 = ax1.bar(data=crp, x='Region', height='Population', color=['r','g','y','b'], edgecolor='black')
p2 = ax2.barh(data=df, y='Country', width='Population', edgecolor='black')


ax1.set_xlabel('Region')
ax1.set_ylabel('Population')
ax1.set_title('Population by Region')
ax1.bar_label(p1, label_type='edge')


ax2.set_xlabel('Population')
ax2.set_ylabel('Countries')
ax2.set_title('Population by Countries')
ax2.bar_label(p2, label_type='edge', padding=5)

plt.show()
```
