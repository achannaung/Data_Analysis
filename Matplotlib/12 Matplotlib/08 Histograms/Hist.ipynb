# Histogram
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
mpl.style.use('ggplot')
```

```python
df = pd.read_csv('players_21.csv')
df.head()
```

```python
len(df)
```

```python
df.shape
```

```python
fig = plt.figure(figsize=(5,5))

ax1 = fig.add_axes([0,0,1,1])
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Chart Title')

ax1.hist(data=df, x='overall', color='cyan')

plt.show()
```

### Changing bin size to 20

```python
fig = plt.figure(figsize=(5,5))

ax1 = fig.add_axes([0,0,1,1])
ax1.set_xlabel('Overall score(bins)')
ax1.set_ylabel('Number of Players')
ax1.set_title('Chart Title')

ax1.hist(data=df, x='overall', bins=20, edgecolor='black') # change bin-size

plt.show()
```

### Changing edgecolor and bins

```python
fig = plt.figure(figsize=(5,5))

ax1 = fig.add_axes([0,0,1,1])
ax1.set_xlabel('Overall score(bins)')
ax1.set_ylabel('Number of Players')
ax1.set_title('Chart Title')

ax1.hist(data=df, x='overall', bins=[0,25,50,75,100], edgecolor='black')

plt.show()
```

### Adding `xticks` to align with `bins`

```python
fig = plt.figure(figsize=(5,5))

ax1 = fig.add_axes([0,0,1,1])
ax1.set_xlabel('Overall score(bins)')
ax1.set_ylabel('Number of Players')
ax1.set_title('Chart Title')

plt.xticks([0,25,50,75,100],[0,25,50,75,100])

ax1.hist(data=df, x='overall', bins=[0,25,50,75,100], edgecolor='black')

plt.show()
```

### Horizontal Alignment

```python
fig = plt.figure(figsize=(5,5))

ax1 = fig.add_axes([0,0,1,1])
ax1.set_ylabel('Overall score(bins)')
ax1.set_xlabel('Number of Players')
ax1.set_title('Chart Title')

plt.yticks([0,25,50,75,100],[0,25,50,75,100])

ax1.hist(data=df, x='overall', bins=[0,25,50,75,100], edgecolor='black',
        orientation='horizontal')

plt.show()
```
