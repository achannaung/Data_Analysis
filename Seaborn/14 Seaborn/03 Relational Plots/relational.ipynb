```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import datetime as dt
```

```python
stocks = pd.read_excel('stocks.xlsx')
stocks.head()
```

```python
stocks['Year'] = stocks['Date'].dt.strftime('%Y')
```

```python
sns.set_theme()
```

```python
# Creating a Line Plot on Seaborn

sns.lineplot(x='Date', y='Close', data=stocks, hue='Symbol')
```

```python
plt.figure(figsize=(20,10))

sns.lineplot(x='Date', y='Close', data=stocks, hue='Symbol',
            linestyle='--',
            palette=['red','blue','lime'])

plt.legend(loc='center left')
plt.show()
```

```python
dfs = stocks[stocks['Symbol']== 'AAPL']
dfs.head(2)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1,3 , figsize=(20,5))

sns.lineplot(x='Date', y='Close', data=dfs, hue='Symbol', ax=ax1,
            linestyle='--', palette=['red'])

sns.lineplot(x='Date', y='Close', data=dfs, hue='Symbol', ax=ax2,
            linestyle='dotted', palette=['blue'])

sns.lineplot(x='Date', y='Close', data=dfs, hue='Symbol', ax=ax3,
            linestyle='dashed', palette=['lime'])
```

## relplot

```python
sns.relplot(x='Date', y='Close', data=stocks, hue='Symbol', 
           kind='line')
```

```python
sns.relplot(x='Date', y='Close', data=stocks, hue='Symbol', kind='line',
           height=3, aspect=3.5,
           row='Symbol')
```
