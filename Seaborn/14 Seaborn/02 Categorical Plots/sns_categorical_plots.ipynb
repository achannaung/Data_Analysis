## Seaborn
### Categorical_Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('top_10_countries.csv')
df.head()
```

```python
sns.barplot(x=df['Region'], y=df['Population'], ci=None    )
```

```python
sns.barplot(y=df['Region'], x=df['Population'],
           orient='h', errorbar=None)
```

```python
sns.barplot(x='Region', y='Population' ,data=df, errorbar=None)
```

```python
sns.barplot(x='Region', y='Population' ,data=df, errorbar=None,
           estimator=sum, hue='Region', dodge=False)

plt.legend(loc='center')
plt.show()
```

```python
sns.barplot(x='Country', y='Population' ,data=df, errorbar=None,
           estimator=sum, hue='Region', dodge=False)

plt.legend(loc='center')
plt.xticks(rotation=90)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(figsize=(18,10), nrows=1, ncols=2, tight_layout=True)

sns.barplot(x='Region', y='Population' ,data=df, errorbar=None, ax=ax1,
           estimator=sum, hue='Region', dodge=False)

sns.barplot(x='Country', y='Population' ,data=df, errorbar=None, ax=ax2,
           estimator=sum, hue='Region', dodge=False)


plt.show()
```
